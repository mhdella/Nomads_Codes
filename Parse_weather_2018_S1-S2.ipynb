```python
numdays = 397
base = datetime.datetime.strptime("2017-12-31 06:00", '%Y-%m-%d %H:%M')
date_list = [base +datetime.timedelta(days=x) for x in range(0, numdays)]
date_list
```

```python
# fout=open("Combine_outs.csv","a")
fout=open("Combine_outs.csv","w")

k=0
missing_data={}

# first file:
for line in open(date_list[0].strftime(f'%Y%m%dT0600.out')+".csv"):
    fout.write(line)
    
# now the rest:  
for num in range(1,len(date_list)):
    try:
        f = open(date_list[num].strftime(f'%Y%m%dT0600.out')+".csv")
    except FileNotFoundError:
        missing_data[k]=date_list[num].strftime(f'%Y%m%dT0600.out')
        k=k+1
        print("oh no a exception")
        continue 
    f.__next__()# skip the header
    f.__next__()# skip the new first forecast overlapped with the forecast of the previous initialization
    f.__next__()# skip the new second forecast overlapped with the forecast of the previous initialization
    f.__next__()# skip the new third forecast overlapped with the forecast of the previous initialization
    
    for line in f:
        fout.write(line)
    f.close() # not really needed
        
fout.close()
```
